# **4. Cleaning the weather data**

Since there are many NaN values before column names (rows 0-7) we need to skip these columns when reading the data.

```python
weather_data = pd.read_excel("/content/gdrive/MyDrive/dokumentasi/cuaca/2016/01. Cuaca-Jakarta-Bulan-Januari-Tahun-2016.xlsx",skiprows=[0,1,2,3,4,5,6,7])
weather_data
```

Because we only need the Temperature,	Humidity,	Precipitation,	Wind Speed, and	Wind Direction data from the dataset. So we need to drop the others.

```python
weather_data.drop(['Tn', 'Tx', 'ss', 'ff_x', 'ddd_x'],axis=1, inplace=True)
weather_data
```

Since there is a lot of unused NaN data at the end of the data, we can use dropna to delete it.

```python
weather_data.dropna(inplace=True)
weather_data
```

Now the weather data is ready to use.

# **5. Combine weather and air quality data**

```python
#load air quality data
aq_data = pd.read_csv("/content/gdrive/MyDrive/dokumentasi/dataset/2016/01. ISPU Provinsi-DKI-Jakarta-Bulan-Januari-Tahun-2016.csv")
#concat air quality and weather data
result = pd.concat([weather_data,aq_data], axis=1)
result
```

Delete unused and duplicates column and rename some column

```python
result.drop(['tanggal', 'critical', 'categori', 'lokasi_spku','max'], axis=1, inplace=True)
result.rename(columns = {'Tanggal':'Date','Tavg': 'Temperature', 'RH_avg': 'Humidity', 'RR' : 'Precipitation', 'ff_avg': 'Wind Speed', 'ddd_car': 'Wind Direction'}, inplace = True)
result
```

Now we succesfully combine the data

# **6. Implement to all of the data and save it to google drive**

```python
#make folder to save the combined data
for x in range(2010,2022):
  !mkdir /content/gdrive/MyDrive/dokumentasi/dataset_combined/$x #Use your own path
```

```python
month={"Januari": "01", "Februari": "02", "Maret": "03", "April": "04", "Mei": "05", "Juni": "06", "Juli": "07", "Agustus": "08", "September": "09", "Oktober":10, "November":11, "Desember":12}
for x in range(2010,2022):
  path_cuaca = f'/content/gdrive/MyDrive/dokumentasi/cuaca/{x}' # use your path
  files_cuaca = glob.glob(path_cuaca + "/*.xlsx")
  path_ispu = f'/content/gdrive/MyDrive/dokumentasi/dataset/{x}' # use your path
  files_ispu = glob.glob(path_ispu + "/*.csv")

  #sort the data to start from januari
  files_cuaca.sort()
  files_ispu.sort()

  for ispu,cuaca in zip(files_ispu, files_cuaca):
    txt = ispu.split("-")
    print(f"Processing: {x} {txt[4]} ")

   
    df_cuaca = pd.read_excel(cuaca, header=0, skiprows=[0,1,2,3,4,5,6,7])
    df_ispu = pd.read_csv(ispu, index_col=None, header=0)
    try:
      df_cuaca.drop(['Tn', 'Tx', 'ss', 'ff_x', 'ddd_x'],axis=1, inplace=True)
    except:
      pass
    df_cuaca.dropna(inplace=True)
    result = pd.concat([df_cuaca, df_ispu], axis=1)

    result.drop(['tanggal', 'critical', 'categori', 'lokasi_spku','max'], axis=1, inplace=True)
    result = result.rename(columns = {'Tavg': 'Temperature', 'RH_avg': 'Humidity', 'RR' : 'Precipitation', 'ff_avg': 'Wind Speed', 'ddd_car': 'Wind Direction'}, inplace = False)
    if txt[4] in month.keys():
      nama_file = f"{month[txt[4]]}. {txt[4]} {x}.csv"
      path_save = f"/content/gdrive/MyDrive/dokumentasi/dataset_combined/{x}/{nama_file}" #use your own path
      try:
        result.drop("Unnamed: 10", axis=1, inplace =True) #Drop unnamed column
      except:
        pass
      result.to_csv(path_save,index=False)
```

# **7. Combine all data to one csv file**

```python
li = []
for x in range(2010,2022):
  path = f'/content/gdrive/MyDrive/dokumentasi/dataset_combined/{x}' # use your path
  all_files = glob.glob(path + "/*.csv")
  all_files.sort()
  for filename in all_files:
      df = pd.read_csv(filename, index_col=None, header=0)
      li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
frame
```

```python
#Check is there's NaN value
frame.isnull().values.any()
```

```python
#Drop NaN value
frame.dropna(inplace=True)
frame.isnull().values.any()
```
