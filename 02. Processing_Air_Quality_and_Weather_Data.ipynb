# **4. Cleaning the weather data**

Since there are many NaN values before column names (rows 0-7) we need to skip these columns when reading the data.

```python
weather_data = pd.read_excel("/content/gdrive/MyDrive/dokumentasi/cuaca/2016/01. Cuaca-Jakarta-Bulan-Januari-Tahun-2016.xlsx",skiprows=[0,1,2,3,4,5,6,7])
weather_data
```

Because we only need the Temperature,	Humidity,	Precipitation,	Wind Speed, and	Wind Direction data from the dataset. So we need to drop the others.

```python
weather_data.drop(['Tn', 'Tx', 'ss', 'ff_x', 'ddd_x'],axis=1, inplace=True)
weather_data
```

Since there is a lot of unused NaN data at the end of the data, we need to drop it.

```python
#make folder to save the combined data
for x in range(2010,2022):
  !mkdir /content/gdrive/MyDrive/dokumentasi/cuaca_cleaned/$x #Use your own path
```

```python
month={"Januari": 31, "Februari": 28, "Maret": 31, "April": 30, "Mei": 31, "Juni": 30, "Juli": 31, "Agustus": 31, "September": 30, "Oktober":31, "November":30, "Desember":30}
for x in range(2010,2021):
  path_cuaca = f'/content/gdrive/MyDrive/dokumentasi/cuaca/{x}' # use your path
  files_cuaca = glob.glob(path_cuaca + "/*.xlsx")
  #sort the data to start from januari
  files_cuaca.sort()

  for cuaca in (files_cuaca):
    txt = cuaca.split("-")
    print(f"Processing: {x} {txt[3]} ")
    df_cuaca = pd.read_excel(cuaca, header=0, skiprows=[0,1,2,3,4,5,6,7])
    #Delete NaN at spesific rows (different for each month)
    if month[txt[3]] == 31:
      df_cuaca.drop(df_cuaca.index[31:47],inplace=True)
    elif month[txt[3]] == 30:
      df_cuaca.drop(df_cuaca.index[30:46],inplace=True)
    elif month[txt[3]] == 28 and x%4 != 0:
      df_cuaca.drop(df_cuaca.index[28:44],inplace=True)
    else:
      df_cuaca.drop(df_cuaca.index[29:45],inplace=True)
    try:
      df_cuaca.drop(['Tn', 'Tx', 'ss', 'ff_x', 'ddd_x'],axis=1, inplace=True)
    except:
      pass
    path_save= cuaca.replace("cuaca", "cuaca_cleaned").replace("xlsx", "csv")
    df_cuaca.to_csv(path_save, index=False)
```

Now the weather data is ready to use.

# **7. Implement to all of the data and save it to google drive**

```python
#make folder to save the combined data
for x in range(2010,2022):
  !mkdir /content/gdrive/MyDrive/dokumentasi/dataset_combined/$x #Use your own path
```

```python
month={"Januari": "01", "Februari": "02", "Maret": "03", "April": "04", "Mei": "05", "Juni": "06", "Juli": "07", "Agustus": "08", "September": "09", "Oktober":10, "November":11, "Desember":12}
for x in range(2010,2021):
  path_cuaca = f'/content/gdrive/MyDrive/dokumentasi/cuaca_cleaned/{x}' # use your path
  files_cuaca = glob.glob(path_cuaca + "/*.csv")
  path_ispu = f'/content/gdrive/MyDrive/dokumentasi/dataset/{x}' # use your path
  files_ispu = glob.glob(path_ispu + "/*.csv")

  #sort the data to start from januari
  files_cuaca.sort()
  files_ispu.sort()

  for ispu,cuaca in zip(files_ispu, files_cuaca):
    txt = ispu.split("-")
    print(f"Processing: {x} {txt[4]} ")

   
    df_cuaca = pd.read_csv(cuaca, header=0)
    df_ispu = pd.read_csv(ispu, index_col=None, header=0)
    result = pd.concat([df_cuaca, df_ispu], axis=1)

    result.drop(['tanggal', 'categori', 'lokasi_spku','max', 'critical'], axis=1, inplace=True)
    result = result.rename(columns = {'Tanggal':'Date','Tavg': 'Temperature', 'RH_avg': 'Humidity', 'RR' : 'Precipitation', 'ff_avg': 'Wind Speed', 'ddd_car': 'Wind Direction'}, inplace = False)
    if txt[4] in month.keys():
      nama_file = f"{month[txt[4]]}. {txt[4]} {x}.csv"
      path_save = f"/content/gdrive/MyDrive/dokumentasi/dataset_combined/{x}/{nama_file}" #use your own path
      try:
        result.drop("Unnamed: 10", axis=1, inplace =True) #Drop unnamed column
      except:
        pass
      result['Precipitation'] = result['Precipitation'].replace([8888.0],0)
      result.to_csv(path_save,index=False)
```

# **8. Combine all data to one csv file**

```python
li = []
for x in range(2010,2021):
  path = f'/content/gdrive/MyDrive/dokumentasi/dataset_combined/{x}' # use your path
  all_files = glob.glob(path + "/*.csv")
  all_files.sort()
  for filename in all_files:
      df = pd.read_csv(filename, index_col=None, header=0)
      li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
frame
```

```python
#Checking non-null and Data type of each column
frame.info()
```

```python
#summarising number of missing values in each column
frame.isnull().sum()
```

```python
#Drop row with NaN at Date column
frame.dropna(subset=["Date"], inplace=True)
```

```python
#Check the result
frame.isnull().sum()
```

```python
#replace unmeasured data with NaN
frame.so2 = frame.so2.replace({"---": np.nan})
frame.co = frame.co.replace({"---": np.nan})
```

```python
#Check data types of each column after replacing with NaN
frame.info()
```

```python
#Convert object data types to numeric
frame["so2"] = pd.to_numeric(frame["so2"])
frame["co"] = pd.to_numeric(frame["co"])
frame.info()
```

```python
#Imputing Missing Values
frame['Temperature'].fillna(frame['Temperature'].mean(), inplace=True)
frame['Humidity'].fillna(frame['Humidity'].mean(), inplace=True)
frame['Precipitation'].fillna(0, inplace=True)
frame['Wind Speed'].fillna(frame['Wind Speed'].mean(), inplace=True)
frame['pm10'].fillna(frame['pm10'].mean(), inplace=True)
frame['so2'].fillna(frame['so2'].mean(), inplace=True)
frame['no2'].fillna(frame['no2'].mean(), inplace=True)
frame['co'].fillna(frame['co'].mean(), inplace=True)
frame['o3'].fillna(frame['o3'].mean(), inplace=True)
```

```python
#Check the result
frame.isnull().sum()
```

```python
#Finding the most common Wind Direction value to fill NaN.
frame["Wind Direction"].value_counts()
```

```python
#Fill NaN.
frame['Wind Direction'].fillna("N", inplace=True)
frame.isnull().sum()
```
